```python
stats = make_array()
for i in np.arange(1000):
    # creates a bootstrap resample
    bootstrap_resample = covid_sample.sample()
    #calculate the statistic on the bootstrap resample (one of the many estimates)
    stat = 100*(1 - (bootstrap_resample.where('DATE_DIED','9999-99-99').num_rows / bootstrap_resample.num_rows)) 
    # death rate statistic stored in 'stats' array
    stats = np.append(stats,stat)
```

```python
table_of_estimates = Table().with_columns('repetitions',np.arange(1000)+1,
                                         'estimated percentage',stats)
table_of_estimates.show()
```

```python
table_of_estimates.sort('estimated percentage').with_column('order# when sorted',np.arange(1000)+1).show()
```

```python
def confidence_interval_lower_b(tbl,level,column_label):
    stats = make_array()
    for i in np.arange(1000):
        stat = np.mean(tbl.sample().column(column_label))
        stats = np.append(stats,stat)
    return percentile( (100 - level)/2 , stats)

def confidence_interval_upper_b(tbl,level,column_label):
    stats = make_array()
    for i in np.arange(1000):
        stat = np.mean(tbl.sample().column(column_label))
        stats = np.append(stats,stat)
    return percentile( 100 - ((100 - level)/2) , stats)
```

```python
column_to_look_at = 'AGE'
level = 95
print('The confidence interval for',column_to_look_at) 
print('is from',confidence_interval_lower_b(covid_sample,level,column_to_look_at))
print('up to',confidence_interval_upper_b(covid_sample,level,column_to_look_at))
```
