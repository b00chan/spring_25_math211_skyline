# Bootstrapping Samples Demo

```python
num_brown = 4
num_blue = 29
num_yellow = 19
num_green = 14
num_orange = 29
num_red = 11

m_m_bag = Table().with_columns('Colors',make_array('Brown','Blue','Yellow','Green','Orange','Red'),
                               'Count (from pic)',make_array(num_brown,num_blue,num_yellow,num_green,num_orange,num_red))
m_m_bag
```

```python
color_list = (['Brown']*num_brown + ['Blue']*num_blue + ['Yellow']*num_yellow + 
              ['Green']*num_green + ['Orange']*num_orange + ['Red']*num_red)

first_bag = Table().with_column('Color',color_list)
first_bag.show()
```

```python
first_bag.sample(first_bag.num_rows,with_replacement = True)
```
